```python
import os
import pandas as pd
list_of_files = os.listdir('data/practic')
user_file = pd.DataFrame()
for file in list_of_files:
    x = 'data/practic/'+ file
    user_file = pd.concat([user_file, pd.read_csv(x)], ignore_index=True)
user_file.drop_duplicates(ignore_index=True)
```

```python
ratings_movies = pd.read_csv(f'data/practic/ratings_movies.csv')

import re
def get_year_release(arg):
    candidates = re.findall(r'\(\d{4}\)', arg)
    if len(candidates) > 0:
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        return None

ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)
ratings_movies['year_release'].isna().sum()
```

```python
ratings_movies[ratings_movies['year_release'] == 1999].groupby('title')['rating'].mean().sort_values()
```

```python
ratings_movies[ratings_movies['year_release'] == 2010].groupby('genres')['rating'].mean().sort_values()
```

```python
ratings_movies.groupby('userId')['genres'].nunique().sort_values(ascending=False)
```

```python
ratings_movies.groupby('userId')['rating'].agg(['count', 'mean']).sort_values(by=['count', 'mean'], ascending=[True, False])
```

```python
ratings_movies[ratings_movies['year_release'] == 2018].groupby('genres')['rating'].agg(['mean', 'count']).sort_values(by=['count', 'mean'], ascending=False)
```

```python
ratings_movies['date'] = pd.to_datetime(ratings_movies['date'])
```

```python
ratings_movies['year_rating'] = ratings_movies['date'].dt.year
```

```python
pivot_movies = ratings_movies.pivot_table(
    values='rating',
    index='year_rating',
    columns='genres',
    aggfunc='mean',
    fill_value=0
)
pivot_movies['Action|Adventure'].sort_values()
pivot_movies['Action|Adventure|Animation|Children|Comedy|IMAX'].sort_values(ascending=False)
pivot_movies['Animation|Children|Mystery'][2018]
pivot_movies['Comedy'].sort_values(ascending=False)
```

```python
orders = pd.read_csv(f'data/practic/orders.csv', sep=';')
display(orders.head(3))
display(orders.shape[0])
products = pd.read_csv(f'data/practic/products.csv', sep=';')
display(products.head(3))
display(products.shape[0])
```

```python
merged_orders = orders.merge(
    products,
    how='inner',
    left_on='ID товара',
    right_on='Product_ID'
)
merged_orders
```
